## Task 1

## Numpy

```python
import matplotlib.pyplot as plt
import numpy as np

pengeluaran = df['Pengeluaran_USD']

# NumPy to automatically set bins limit
bin_edges = np.histogram_bin_edges(pengeluaran, bins='auto')

# histogram
plt.hist(pengeluaran, bins=bin_edges, edgecolor='k')
plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran Menggunakan NumPy')
plt.show()
```

## Bins using list

```python
# Using bin constraint list
custom_bins = [0, 20, 40, 60, 80, 100]

# Create a histogram with custom bin boundaries
plt.hist(pengeluaran, bins=custom_bins, edgecolor='k')
plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran dengan Batasan Bin Kustom')
plt.show()
```

## Multiple histogram

```python
# histogram untuk setiap kelompok data
plt.hist(df[df['Gender'] == 'Pria']['Pengeluaran_USD'], bins=10, edgecolor='k', alpha=0.5, label='Pria')
plt.hist(df[df['Gender'] == 'Wanita']['Pengeluaran_USD'], bins=10, edgecolor='k', alpha=0.5, label='Wanita')

plt.xlabel('Pengeluaran (USD)')
plt.ylabel('Frekuensi')
plt.title('Histogram Pengeluaran untuk Pria dan Wanita')
plt.legend()
plt.show()
```
